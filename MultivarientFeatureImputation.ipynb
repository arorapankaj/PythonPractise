```python
csv_path = os.path.join('datasets', 'data-prep')
diabetes_csv = os.path.join(csv_path, 'diabetes_processed_incomplete.csv')
diabetes = pd.read_csv(diabetes_csv)
diabetes.sample(10)
```

```python
diabetes.isnull().sum() #Indulin was remaining
```

```python
diabetes_features = diabetes.drop(['Outcome'], axis=1)
diabetes_label = diabetes[['Outcome']]
```

```python
diabetes_features.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter=10000, random_state=0)
```

```python
imp.fit(diabetes_features)
```

```python
diabetes_features_arr = imp.transform(diabetes_features)
```

```python
diabetes_features_arr.shape
```

```python
diabetes_features = pd.DataFrame(diabetes_features_arr, columns=diabetes_features.columns)
diabetes_features.head()
```

```python
diabetes = pd.concat([diabetes_features, diabetes_label], axis=1)
diabetes.sample(5)
```

```python
diabetes.head()
```

```python
diabetes.isnull().sum()
```

```python
diabetes.to_csv(os.path.join(csv_path, 'diabetes_processed.csv'), index=False)
```
